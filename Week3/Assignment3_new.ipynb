## Autonomous systems AS-733 network

"The graph of routers comprising the Internet can be organized into sub-graphs called Autonomous Systems (AS)

Each Autonomous system exchanges traffic flows with some neighbors (peers). The dataset contains 733 daily instances which span an interval of 785 days from November 8 1997 to January 2 2000."

Datasource: https://snap.stanford.edu/data/as-733.html

```python


import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

g=nx.read_edgelist("https://raw.githubusercontent.com/nobieyi00/CUNY-SPS-DATA620/master/Week3/as20000102.txt")

nx.draw(g)
```

Display the number of edges

```python
g.number_of_edges()
```

Display the number of nodes

```python
g.number_of_nodes()
```

Calculate the average shortest path


```python

from networkx import algorithms
algorithms.average_shortest_path_length(g)
```

Graph diameter
A graph's diameter is the largest number of vertices that must be traversed in order to
travel from one vertex to another. Let's find it

```python
#diameter
nx.diameter(g)
```
