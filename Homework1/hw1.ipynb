In this homework, I will design a graph network using NetworkX package


First step is to import the NetworkX package

```python
import networkx as net
```

Instantitate Graph

```python
o=net.Graph()
```

Add Edges

```python
o.add_edge('Jane','Ike')
o.add_edge('Ike','Heather')
o.add_edge('Heather','Garth')
o.add_edge('Heather','Fernando')
o.add_edge('Garth','Ed')
o.add_edge('Garth','Diane')
o.add_edge('Garth','Beverly')
o.add_edge('Garth','Fernando')
o.add_edge('Ed','Diane')
o.add_edge('Ed','Beverly')
o.add_edge('Beverly','Diane')
o.add_edge('Beverly','Andre')
o.add_edge('Andre','Diane')
o.add_edge('Andre','Fernando')
o.add_edge('Andre','Carol')
o.add_edge('Carol','Diane')
o.add_edge('Carol','Fernando')
o.add_edge('Fernando','Diane')
```

We now draw the network

```python
net.draw(o,with_labels=True)
```

Another way to accomplish this with cleaner code is to use dictionary data structure

```python
edges = {"Jane" : {"Ike"}, 
         "Ike" : {"Jane", "Heather"}, 
         "Heather": {"Ike", "Fernando", "Garth"},
         "Fernando": {"Heather", "Garth", "Diane", "Andre", "Carol"},
         "Garth": {"Heather", "Fernando", "Diane", "Beverly", "Ed"},
         "Diane": {"Fernando", "Garth", "Carol", "Andre", "Beverly", "Ed"},
         "Carol": {"Fernando", "Diane", "Andre"},
         "Andre": {"Fernando", "Diane", "Carol", "Beverly"},
         "Beverly": {"Andre", "Diane", "Garth", "Ed"},
         "Ed": {"Beverly", "Diane", "Garth"}
        }
```

```python
G = net.from_dict_of_lists(edges)
```

```python
net.draw(G, with_labels=True)
```
